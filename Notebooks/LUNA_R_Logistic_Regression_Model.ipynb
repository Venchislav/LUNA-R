```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""

 __      __     _      _____ _____       _______ ____  _____
 \ \    / /\   | |    |_   _|  __ \   /\|__   __/ __ \|  __ \
  \ \  / /  \  | |      | | | |  | | /  \  | | | |  | | |__) |
   \ \/ / /\ \ | |      | | | |  | |/ /\ \ | | | |  | |  _  /
    \  / ____ \| |____ _| |_| |__| / ____ \| | | |__| | | \ \
     \/_/    \_\______|_____|_____/_/    \_\_|  \____/|_|  \_\


                          VALIDATOR V 0.01
                              --------

          InputValidation class Purpose is to validate
          inputs of particular model. It raises error messages
          with help on fixing this bug.

          Conditions checked:
          1. Data Shapes
          2. Data Types
          3. Appropriate argument values

"""
class InputValidation:
  def __init__(self):
    pass

  def logistic_regression_valid(self, X, y, epochs, lr, verbose, loss):
    try:
      assert(X.shape[0] == y.shape[0])
    except:
      raise ValueError("X shape doesn't correspond to y shape  (number of samples doesn't match number of labels)")

    assert(loss.lower() in self.supported_loss)

    assert(type(epochs) == int)
    assert(type(lr) == float)

    # if everything is fine
    return 0
```

```python
"""
  __  __  ____  _____ _____ ______ _____ ______ _____   _____
 |  \/  |/ __ \|  __ \_   _|  ____|_   _|  ____|  __ \ / ____|
 | \  / | |  | | |  | || | | |__    | | | |__  | |__) | (___
 | |\/| | |  | | |  | || | |  __|   | | |  __| |  _  / \___ \
 | |  | | |__| | |__| || |_| |     _| |_| |____| | \ \ ____) |
 |_|  |_|\____/|_____/_____|_|    |_____|______|_|  \_\_____/


                   MODIFIERS V 0.01
                       --------

              Modifiers are used for fitting
              process customization and control.

              V 0.01 Modifier List:
              1. EarlyStopping
"""


class Modifiers:
  def __init__(self):
    pass

  def early_stopping(loss, history, patience, temp_patience, delta):
    history = [float('inf')] + history
    if min(history) - loss >= delta:
      return 1

    elif temp_patience >= patience:
      return 0

    else:
      temp_patience += 1
    return 2
```

```python

"""

  _      ____   _____ _____ ______  _____
 | |    / __ \ / ____/ ____|  ____|/ ____|
 | |   | |  | | (___| (___ | |__  | (___
 | |   | |  | |\___ \\___ \|  __|  \___ \
 | |___| |__| |____) |___) | |____ ____) |
 |______\____/|_____/_____/|______|_____/


             LOSS MODULE V 0.01
                 --------

    Losses class Purpose is to set
    different cost functions for
    different models. Via these
    cost functions model will be optimized
    with optimization function.

    V 0.01 Cost Function List:
    1. Mean Squared Error (mse)
    2. Mean Absolute Error (mae)

"""


class Losses:
  def __init__(self):
    pass

  # ----------REGRESSION----------
  def mean_squared_error(self, y_true, y_pred):
    loss = 0

    for i in range(len(y_true)):
      loss += (y_true[i] - y_pred[i]) ** 2

    loss = loss / len(y_true)

    return loss


  def mean_absolute_error(self, y_true, y_pred):
    loss = 0

    for i in range(len(y_true)):
      loss += abs(y_true[i] - y_pred[i])

    loss = loss / len(y_true)

  # ----------CLASSIFICATION----------


  def logistic_cost(self, y_pred, y):
    cost = 0

    for i in range(len(y_pred)):
      if y[i] == 1:
        cost = cost - np.log(y_pred[i])
      elif y[i] == 0:
        cost = cost - np.log(1 - y_pred[i])
      else:
        raise ValueError("WTF DUDE?! Seems you need something else")

    return 1 / len(y_pred) * cost
```

```python

"""
  _      ____   _____ _____  _____ _______ _____ _____   _____  ______ _____ _____  ______  _____ _____ _____ ____  _   _
 | |    / __ \ / ____|_   _|/ ____|__   __|_   _/ ____| |  __ \|  ____/ ____|  __ \|  ____|/ ____/ ____|_   _/ __ \| \ | |
 | |   | |  | | |  __  | | | (___    | |    | || |      | |__) | |__ | |  __| |__) | |__  | (___| (___   | || |  | |  \| |
 | |   | |  | | | |_ | | |  \___ \   | |    | || |      |  _  /|  __|| | |_ |  _  /|  __|  \___ \\___ \  | || |  | | . ` |
 | |___| |__| | |__| |_| |_ ____) |  | |   _| || |____  | | \ \| |___| |__| | | \ \| |____ ____) |___) |_| || |__| | |\  |
 |______\____/ \_____|_____|_____/   |_|  |_____\_____| |_|  \_\______\_____|_|  \_\______|_____/_____/|_____\____/|_| \_|




                                        Logistic Regression Model V 0.01
                                                  --------

                                        Logistic Regression great classification
                                        algorithm that is widely used.

                                        This class implements Logistic Regression
                                        with Gradient Descent optimization.


                                                  -------

                                              M E T H O D S:

                                                  __init__

                                            Basic Initialization

                                            - loss history init
                                            - modifiers list init
                                            - supported losses init


                                                    fit


                                            - Fitting process algorithm
                                                  -------
                                                    fit
                                                I N P U T:

                                              - array-like data
                                              - number of epochs
                                              - learning_rate (0.01 by default)
                                              - verbose level
                                                (0 - no output;
                                                1 - Only number of epochs;
                                                2 and more - Full info with loss)
                                              - loss function (crossentropy by default)

                                                  -------
                                                    fit
                                              O U T P U T:

                                              - weights
                                              - bias

                                        V 0.01 Cost Function List:
                                        1. Crossentropy


"""



class LogisticRegression:
  def __init__(self):

    # useful variables initialization:

    self.loss_history = []
    self.supported_modifiers = {'early_stopping_modifier': Modifiers.early_stopping}
    self.supported_loss = {'logistic_cost': Losses.logistic_cost}


  def fit(self, X, y, epochs=0, lr=0.01, verbose=2, loss_function='logistic_cost', modifier=None, delta=0, patience=0, reg=1):
    # catch errors:
    InputValidation.logistic_regression_valid(self, X, y, epochs, lr, verbose, loss_function)

    temp_patience = 0

    n_samples, n_features = X.shape
    self.weights = np.zeros(n_features)
    self.bias = 0

    for epoch in range(epochs):
      dw = np.zeros(n_features)
      db = 0

      for i in range(n_samples):
        dw += (1 / (1 + np.e ** (-np.dot(self.weights, X[i]) + self.bias)) - y[i]) * X[i] + (reg / n_samples) * self.weights
        db += 1 / (1 + np.e ** (-np.dot(self.weights, X[i]) + self.bias)) - y[i]

      self.weights -= 1 / n_samples * dw * lr
      self.bias -= 1 / n_samples * db * lr

      ### stats for nerds🤓

      loss = self.supported_loss[loss_function](self, np.array(self.predict(X)), y)
      if verbose >= 2:
        print(f'On Epoch: {epoch} | Loss is: {loss}')
      elif verbose == 1:
        print(f'Epoch: {epoch}/{epochs}')

      # modifiers

      if modifier:
        if modifier in self.supported_modifiers:
          result = self.supported_modifiers[modifier](loss, self.loss_history, patience, temp_patience, delta)
          if result == 1:
            temp_patience = 0

          elif result == 2:
            temp_patience += 1

          else:
            return self.weights, self.bias

      self.loss_history.append(loss)

    return self.weights, self.bias


  def predict(self, X):
    preds = []

    for x_i in X:
      preds.append(1 / (1 + np.e ** (-np.dot(x_i, self.weights) + self.bias)))

    return preds
```
